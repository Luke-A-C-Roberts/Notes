```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im1 = cv2.imread("/home/luke/Downloads/1_pxSk8L4untWMRhRLBfL6KQ-1995556012.png")
im2 = cv2.imread("/home/luke/Downloads/ijgi-10-00666-g002a-1523168004.png")
im3 = cv2.imread("/home/luke/Downloads/th-1107877450")
im4 = cv2.imread("/home/luke/Downloads/th-1303226680")
```

```python
def process_image(im):
    return cv2.normalize(im, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX) // 32 * 32
```

```python
im1 = process_image(im1)
im2 = process_image(im2)
im3 = process_image(im3)
im4 = process_image(im4)
```

```python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 20))

ax1.imshow(im1)
ax2.imshow(im2)
ax3.imshow(im3)
ax4.imshow(im4)
```
